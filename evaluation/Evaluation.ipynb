```python
# Load all OCR Texts, Ground Truth, Fill_Mask_Corrected Texts, llama2-13b corrected texts
import pickle

BLN600_texts_dict = pickle.load(open('../datasets/BLN600/BLN600.pkl', 'rb'))
fs_corrected_texts = pickle.load(open('../datasets/BLN600/fs_corrected.pkl', 'rb'))
llama2_13b_corrected_texts = pickle.load(open('../datasets/BLN600/llama2_13b_corrected.pkl', 'rb'))
gpt4o_corrected_texts = pickle.load(open('../datasets/BLN600/GPT4o_corrected.pkl', 'rb'))
```

```python
ocr_texts = BLN600_texts_dict['ocr']
ground_truth = BLN600_texts_dict['ground_truth']
```

```python
import pandas as pd

all_texts_dict = []
for filename in gpt4o_corrected_texts:
    sample_id = int(filename[:-4])
    if sample_id not in llama2_13b_corrected_texts:
        continue
    all_texts_dict.append({
        "id": sample_id,
        "ocr": ocr_texts[filename],
        "ground_truth": ground_truth[filename],
        "fs_corrected": fs_corrected_texts[filename],
        "llama2_13b_corrected": llama2_13b_corrected_texts[sample_id],
        "gpt4o_corrected": gpt4o_corrected_texts[filename],
    })
all_texts_df = pd.DataFrame(all_texts_dict)
```

```python
# normalise all texts (tokenize and rejoin with space)
from postcorrection.utils import tokenize_text
all_texts_df["ocr"] = all_texts_df["ocr"].apply(lambda x: ' '.join(tokenize_text(x)))
all_texts_df["ground_truth"] = all_texts_df["ground_truth"].apply(lambda x: ' '.join(tokenize_text(x)))
all_texts_df["llama2_13b_corrected"] = all_texts_df["llama2_13b_corrected"].apply(lambda x: ' '.join(tokenize_text(x)))
all_texts_df["gpt4o_corrected"] = all_texts_df["gpt4o_corrected"].apply(lambda x: ' '.join(tokenize_text(x)))
```

```python
# add cer
from evaluation.util import get_cer
all_texts_df['ocr_cer'] = all_texts_df.apply(lambda row: get_cer(row['ocr'], row['ground_truth']), axis=1)
all_texts_df['fs_cer'] = all_texts_df.apply(lambda row: get_cer(row['fs_corrected'], row['ground_truth']), axis=1)
all_texts_df['llama2_13b_cer'] = all_texts_df.apply(lambda row: get_cer(row['llama2_13b_corrected'], row['ground_truth']), axis=1)
all_texts_df["gpt4o_cer"] = all_texts_df.apply(lambda row: get_cer(row['gpt4o_corrected'], row['ground_truth']), axis=1)
```

```python
# add cer reduction
all_texts_df['fs_cer_reduction'] = all_texts_df.apply(lambda row: ((row['ocr_cer'] - row['fs_cer']) / row['ocr_cer']) * 100, axis=1)
all_texts_df['llama2_13b_cer_reduction'] = all_texts_df.apply(lambda row: ((row['ocr_cer'] - row['llama2_13b_cer']) / row['ocr_cer']) * 100, axis=1)
all_texts_df['gpt4o_cer_reduction'] = all_texts_df.apply(lambda row: ((row['ocr_cer'] - row['gpt4o_cer']) / row['ocr_cer']) * 100, axis=1)
```

```python
# add word error rate
import jiwer
all_texts_df['ocr_wer'] = all_texts_df.apply(lambda row: jiwer.wer(row['ocr'], row['ground_truth']), axis=1)
all_texts_df['fs_wer'] = all_texts_df.apply(lambda row: jiwer.wer(row['fs_corrected'], row['ground_truth']), axis=1)
all_texts_df['llama2_13b_wer'] = all_texts_df.apply(lambda row: jiwer.wer(row['llama2_13b_corrected'], row['ground_truth']), axis=1)
all_texts_df['gpt4o_wer'] = all_texts_df.apply(lambda row: jiwer.wer(row['gpt4o_corrected'], row['ground_truth']), axis=1)
```

```python
# add wer reduction
all_texts_df['fs_wer_reduction'] = all_texts_df.apply(lambda row: ((row['ocr_wer'] - row['fs_wer']) / row['ocr_wer']) * 100, axis=1)
all_texts_df['llama2_13b_wer_reduction'] = all_texts_df.apply(lambda row: ((row['ocr_wer'] - row['llama2_13b_wer']) / row['ocr_wer']) * 100, axis=1)
all_texts_df['gpt4o_wer_reduction'] = all_texts_df.apply(lambda row: ((row['ocr_wer'] - row['gpt4o_wer']) / row['ocr_wer']) * 100, axis=1)
```
